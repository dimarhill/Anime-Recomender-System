```python
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
import pickle
```

```python
rating = pd.read_csv('../datasets/rating.csv.zip')
```

```python
anime = pd.read_csv('../datasets/anime.csv')
```

```python
df = pd.merge(rating, anime, on='anime_id', how='inner')
```

```python
dfmovie = df[df['type'] == 'Movie']
```

```python
dftv = df[df['type'] == 'TV']
```

```python
dfog = df[df['type'].isin(['OVA', 'ONA'])]
```

```python
dfog
```

```python
df[df['name'] == 'Mushishi Special: Hihamukage']
```

```python
dfog.head()
```

```python
dfspecial = df[df['type'] == 'Special']
```

```python
dfmovie.drop(columns = ['genre', 'type', 'members', 'episodes', 'rating_y'], inplace=True)
```

```python
dftv.drop(columns = ['genre', 'type', 'members', 'episodes', 'rating_y'], inplace=True)
```

```python
dfog.drop(columns = ['genre', 'type', 'members', 'episodes', 'rating_y'], inplace=True)
```

```python
dfspecial.drop(columns = ['genre', 'type', 'members', 'episodes', 'rating_y'], inplace=True)
```

```python
dfmovie
```

```python
rating_mxmovie = pd.pivot_table(
    dfmovie,
    index='name',
    columns='user_id',
    values='rating_x'
)
```

```python
rating_mxtv = pd.pivot_table(
    dftv,
    index='name',
    columns='user_id',
    values='rating_x'
)

```

```python
rating_mxog = pd.pivot_table(
    dfog,
    index='name',
    columns='user_id',
    values='rating_x'
)

```

```python
rating_mxspecial = pd.pivot_table(
    dfspecial,
    index='name',
    columns='user_id',
    values='rating_x'
)

```

```python
ratings_sparsemovie = sparse.csr_matrix(rating_mxmovie.fillna(0))
```

```python
ratings_sparsetv = sparse.csr_matrix(rating_mxtv.fillna(0))
```

```python
ratings_sparseog = sparse.csr_matrix(rating_mxog.fillna(0))
```

```python
ratings_sparsespecial = sparse.csr_matrix(rating_mxspecial.fillna(0))
```

```python
distsmovie = pairwise_distances(ratings_sparsemovie, metric='cosine')
```

```python
diststv = pairwise_distances(ratings_sparsetv, metric='cosine')
```

```python
distsog = pairwise_distances(ratings_sparseog, metric='cosine')
```

```python
distsspecial = pairwise_distances(ratings_sparsespecial, metric='cosine')
```

```python
df_recmovie = pd.DataFrame(distsmovie, columns=rating_mxmovie.index, index=rating_mxmovie.index)
```

```python
df_rectv = pd.DataFrame(diststv, columns=rating_mxtv.index, index=rating_mxtv.index)
```

```python
df_recog = pd.DataFrame(distsog, columns=rating_mxog.index, index=rating_mxog.index)
```

```python
df_recog
```

```python
df_recspecial = pd.DataFrame(distsspecial, columns=rating_mxspecial.index, index=rating_mxspecial.index)
```

```python
df_recmovie.insert(0, "name", df_recmovie.index)
```

```python
df_rectv.insert(0, "name", df_rectv.index)
```

```python
df_recog.insert(0, "name", df_recog.index)
```

```python
df_recspecial.insert(0, "name", df_recspecial.index)
```

```python
df_recmovie.to_pickle('../datasets/recmovie.plk')
```

```python
df_rectv.to_pickle('../datasets/rectv.plk')
```

```python
df_recog.to_pickle('../datasets/recog.plk')
```

```python
df_recspecial.to_pickle('../datasets/recspecial.plk')
```
